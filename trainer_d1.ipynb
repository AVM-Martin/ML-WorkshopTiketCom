# Introduction Into Pandas

First, import the pandas library

```python
import pandas as pd
```

## Creating Data

There are two core objects in pandas : DataFrame and Series
    
Dataframe is a table (2D Array)

Here is how we create a DataFrame

```python
pd.DataFrame({'A':[1, 0], 'B':[0, 1]})
```

```python
pd.DataFrame({'Student':['Ann', 'Bob'], 'Grade':[100, 75], 'Index':['A', 'B']})
```

We use pd.DataFrame to generate these DataFrame objects.

The list of row labels used in a DataFrame is known as Index. For the above example, the index are 0, 1.
We can assign values to index by using index parameter when we construct a list

```python
pd.DataFrame({'Student':['Ann', 'Bob'], 
              'Grade':[100, 75], 'Index':['A', 'B']},
             index=['Class 1-A', 'Class 1-B']
            )
```

A series is a sequence of a data values (1D Array) and can be seen as a list.

Here is how to create a series

```python
pd.Series([1, 0, 1, 0, 1, 1])
```

A Series is a single column of a DataFrame, so we can assign column values to the Series the same way with how to assign values to a DataFrame, which is using an index parameter.
Notes that a Series do not have a column name, only has one overall name :

```python
pd.Series([0, 20, 40, 60, 80], index=['E', 'D', 'C', 'B', 'A'], name='Grade Lowest Bound')
```

DataFrame can be seen as a bunch of Series that "glued together"

# Data Cleaning

Let's try to load the data again and repeat the data insight step.

You will find out that there are several problem with the data like duplicate data and missing values

```python
data = pd.read_csv('https://raw.githubusercontent.com/davidbrandon16/binusMachineLearning/master/sample.csv', sep=';')
```

```python
data.head()

data.info()
```

First, we wanted to remove some duplicate values using drop_duplicates

```python
data = data.drop_duplicates()
```

```python
data.info()
```

```python
data.head()
```

Notice that in the Value columns, there are only 231 non-null values. That's mean there are about 9 null values in the data. These null values are called missing values. 

There are several ways that we can do with missing values :
1. Delete the rows/columns contained missing values
2. Fill the missing values with certain value (0 or average)

Which one that might be suitable for these? 

For deleting the rows/colums we use dropna() while for fill the missing values we use fillna.

Since we will try to use both of these methods, better duplicate the dataframe first

```python
data2 = data
```

```python
data2.head()
```

Let's drop the rows that contain NaN value on data2 and do some other things with that

```python
data2 = data2.dropna()

print(data2.head())

print(data2.info())
```

Notice that the index is not right because of some rows we dropped before. We can use reset_index

```python
data2 = data2.reset_index(drop=True)

data2.head()
```

We can try to rename the column index using rename

```python
data2.columns
```

```python
data2 = data2.rename(columns={'Industry_aggregation_NZSIOC':'Level', 
                      'Industry_code_NZSIOC':'Code', 
                      'Industry_name_NZSIOC':'Name'})

data2.head()
```

We can try to group the data and find out some statistical computation behind using groupby, followed by one (or more) statistical function

```python
data2.groupby(['Level', 'Name']).mean()
```

Now, we try to change the value of Value column based on Units. If Units = Dollars (millions) then we multiply value with 1000.

```python
data2.loc[data2['Units']=='Dollars (millions)', 'Value'] = data2['Value']*1000

data2.head()
```

Next, we can change the type of the column using astype

```python
data2.Value = data2['Value'].astype('int64')

data2.head()
```

Try to sort the value based on Variable_code

```python
data2 = data2.sort_values(by=['Name', 'Code', 'Variable_code'])

data2.head()

data2 = data2.reset_index(drop=True)

data2.head()
```

We can also drop the unused columns. Let's try to drop Year columns since all the year columns have 2018 value and does not useful. 

```python
data2 = data2.drop(columns='Year')

data2.head()
```

Now let's try another type of handling missing value, which is fill the value using fillna

```python
data3 = data

data3.head()
```

```python
data3 = data3.fillna(0)

data3.head()
```

# Data Visualization

First Import pyplot from matplotlib

```python
import matplotlib.pyplot as plt
```

```python
data4 = data2
```

```python
data4
```

select value only level name and value

```python
data4 = data4[["Level","Name","Value"]]
```

```python
data4
```

```python
data_level_3 = data4[data4.Level == "Level 3"]
data_level_4 = data4[data4.Level == "Level 4"]
```

Get Average from name by each group

```python
data_level_3 = data_level_3.groupby(["Name"]).mean()
data_level_4 = data_level_4.groupby(["Name"]).mean()
```

```python
data_level_3
```

```python
data_level_4
```

Show as plot lib using bar

## Explaination
1. figure -> set size of diagram
2. title set title for graphic
3. bar means show as bar
4. plot show as line
5. show -> display

```python
plt.figure(figsize=(20,10))
plt.title("data Level 3")
plt.bar(data_level_3.index, data_level_3.Value, width = 0.8)
```

```python
plt.figure(figsize=(20,10))
plt.title("Differences between Level")
plt.plot(data_level_3, label="level 3")
plt.plot(data_level_4, label="level 4")
plt.legend()
plt.show()
```
